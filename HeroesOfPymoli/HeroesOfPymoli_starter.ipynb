```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate number of unique items based on Item ID
items_unique=purchase_data['Item ID'].unique()
numberUniqueItems=len(items_unique)

#calculate number of purchases based on Purchase ID
shape_purchaseID=purchase_data.shape
numberOfPurchases=shape_purchaseID[0]

#calculate the average price based on average transcation 
averagePrice=purchase_data['Price'].sum()/numberOfPurchases
averagePrice="${:,.2f}".format(averagePrice)

#calculate total revenue
totalRevenue="${:,.2f}".format(purchase_data['Price'].sum())

#create a summary table
summaryTable=pd.DataFrame({
    "Number of Unique Items":[numberUniqueItems],
    "Average Price":[averagePrice],
    "Number of Purchases":[numberOfPurchases],
    "Total Revenue":[totalRevenue]
})

summaryTable
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#take total purchase data to sort by gender
#simplify the purchase data
newPurchase_data=purchase_data.drop(columns=['Purchase ID', 'Item ID', 'Age', 'Item Name'])
#print(newPurchase_data)
      
#purchase count by gender
genderPurchase=newPurchase_data["Gender"].value_counts()
#print(genderPurchase)
#print(genderPurchase.dtypes)

#total purchases by gender
genderSum=newPurchase_data.groupby("Gender").sum()

#print(genderSum)
#print(genderSum.dtypes)

#average total purchase per person by gender
genderAvgPurchase=newPurchase_data.groupby(["Gender"]).mean()

#avg total purchase per person by gender
#avg total purchase per person = Count of each gender / Total Purchase per Gender
uniqueSN=purchase_data.drop_duplicates(subset="SN")
genderCount=uniqueSN['Gender'].value_counts()
avgTotalPurchase=genderSum['Price']/genderCount

#put information into a summary table
genderTable=pd.DataFrame({ "Purchase Count": genderPurchase.map("{0:,.0f}".format),
                            "Average Purchase Price": genderAvgPurchase['Price'].map("${0:,.2f}".format),
                            "Total Purchase Value": genderSum['Price'].map("${0:,.2f}".format),
                            "Avg Total Purchase Per Person": avgTotalPurchase.map("${0:,.2f}".format)
                         })
genderTable
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#bin purchase data by age range. Include all purchases
age_bins=[0, 9, 14, 19, 24, 29, 34, 39, 200]
age_labels=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_data=purchase_data
age_data['Age Range']=pd.cut(purchase_data['Age'], age_bins, labels=age_labels)

#calculate purchase count by age range
purchaseCount=age_data.groupby(['Age Range']).count()

#calculate average purchase by age range
purchaseMean=age_data.groupby(['Age Range']).mean()

#calculate total purchase value by age range
purchaseTotalValue=age_data.groupby(['Age Range']).sum()

#calculate avg total purchase value by age range
#need to isolate data for each individual player - so their repeated purchases don't skew the count
#use dataframe set that has only unique usernames "uniqueSN"
age_data['Age Range']=pd.cut(uniqueSN['Age'], age_bins, labels=age_labels)
TotalCount=age_data.groupby(['Age Range']).count()
purchaseAvgTotal=purchaseTotalValue/TotalCount

#compile information into summary table
agePurchaseTable=pd.DataFrame({"Purchase Count": purchaseCount["Price"],
                               "Average Purchase Price" : purchaseMean["Price"].map("${0:,.2f}".format),
                               "Total Purchase Value" : purchaseTotalValue["Price"].map("${0:,.2f}".format),
                               "Avg Total Purchase Per Person" : purchaseAvgTotal['Price'].map("${0:,.2f}".format)
                              })
agePurchaseTable
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#order of table based on total purchase value by single user
userPurchase=purchase_data.groupby("SN")

#get total purchase value, sort with highest value on top of table
PurchaseValue=userPurchase.sum()
PurchaseValue=PurchaseValue.sort_values("Price", ascending=False)
totalPurchaseValue=PurchaseValue["Price"]


#find the users with the most total purchase value
#need to reset index for Purchase value
PurchaseValue=PurchaseValue.reset_index()
topUsers=PurchaseValue["SN"]

#calculate average purchase price
avgPurchase=userPurchase.mean()
#merge mean Price with original table where users are in order of total purchase price
mergePurchase=pd.merge(PurchaseValue, avgPurchase, on="SN")
mergePurchase.head()
#clean up table, reduce to username, total price, and avergage price
mergePurchase=mergePurchase.iloc[:, [0,4,8]]
mergePurchase=mergePurchase.rename(columns={'Price_x':'Total Purchase Value', 'Price_y':'Average Purchase Price'})
mergePurchase.head()

#calculate purchase count
countPurchase=userPurchase.count()
#simplify dataframe to single count column for purchases per SN
countPurchase=countPurchase["Purchase ID"]
#merge with summary table and clean up column
mergePurchase=pd.merge(mergePurchase, countPurchase, on="SN")
mergePurchase=mergePurchase.rename(columns={'Purchase ID':'Purchase Count'})
mergePurchase.head()

#change formatting
mergePurchase['Total Purchase Value']=mergePurchase['Total Purchase Value'].map("${0:,.2f}".format)
mergePurchase['Average Purchase Price']=mergePurchase['Average Purchase Price'].map("${0:,.2f}".format)
mergePurchase.head()
```


## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#keep only columsn related to items
itemsData=purchase_data.drop(columns=['Purchase ID', 'SN', 'Gender', 'Age Range'])
itemsData.head()

itemsData_group=itemsData.groupby(['Item ID', 'Item Name'])


#determine most frequently purchased items
itemsDataCount=pd.DataFrame(itemsData_group['Price'].count())
itemsDataCount=itemsDataCount.rename(columns={"Price":"Purchase Count"})
itemsDataCount=itemsDataCount.reset_index()

#determine sum of purchased items
itemsSum=pd.DataFrame(itemsData_group['Price'].sum())
itemsSum=itemsSum.rename(columns={"Price":"Total Purchase Value"})
itemsSum=itemsSum.reset_index()

#create unique items dataframe to merge calcualted groupby dataframes
uniqueItems=purchase_data.drop_duplicates(subset="Item ID")
itemsSum=pd.merge(uniqueItems, itemsSum, on='Item ID', how='inner')

#merge itemsSum and itemsCount to ensure is in the right order. Doing so will mean Purchase Count is not in order
itemsSum=pd.merge(itemsSum, itemsDataCount, on='Item ID', how='inner')

#eliminate excessive columsn from merging multiple dataframes together

itemsSum=itemsSum.drop(columns=['Purchase ID', 'SN', 'Gender', 'Age Range', 'Age', 'Item Name_y'])

#create new dataframe as summary table
#dataframes created by group by are organized differently then in itemsData. The item info will not match to the groupby information
itemsSummary=pd.DataFrame({"Item ID": itemsSum['Item ID'],
                           #when merged into itemSum there were two columns with the name
                            "Item Name": itemsSum['Item Name_x'],
                            "Purchase Count": itemsSum['Purchase Count'],
                            "Item Price" : itemsSum['Price'],
                            "Total Purchase Value": itemsSum['Total Purchase Value']
                          })

#reorganize to show top Total Purchase Value earners
itemsSummary=itemsSummary.sort_values("Purchase Count", ascending=False)
itemsSummary = itemsSummary.reset_index()


#format price data (must do after sorting, because formatted values are not numeric)
itemsSummary['Item Price']=itemsSummary['Item Price'].map("${0:,.2f}".format)
itemsSummary['Total Purchase Value']=itemsSummary['Total Purchase Value'].map("${0:,.2f}".format)

itemsSummary.head()
###missing "Item ID 108, Extraction, Quickblade of Trembling Hands"
```
